# One hot encoding

```python
from sklearn import preprocessing
```

```python
one_enc=preprocessing.OneHotEncoder()
```

```python
featured_df=df.drop(columns=['species'])
```

```python
encoded_species=one_enc.fit_transform(df[['species']])
enc_df=pd.DataFrame(encoded_species.toarray())
enc_df.head()
```

```python
df_encode=featured_df.join(enc_df)
```

```python
df_encode.head()
```

```python
df_encode.rename(columns={0:'setosa',1:'versicular',2:'virginia'},inplace=True)
```

```python
df_encode.head()
```

```python
df_encode['setosa']=df_encode['setosa'].astype('int')
```

```python
df_encode['versicular']=df_encode['versicular'].astype('int')
```

```python
df_encode['virginia']=df_encode['virginia'].astype('int')
```
