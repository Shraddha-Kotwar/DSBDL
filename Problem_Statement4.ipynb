1. Load the demo dataset in dataframe object df

2. Detect the outlier using Z-score

3. replace the outliers with the median value.

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore

df = pd.read_csv(r"C:\Users\OM\Downloads\demo1.csv")
df.head()
```

```python
# 2. Detecting outliers using Z-score
# Computing Z-score for each column
z_scores = np.abs(zscore(df))
z_scores

# Threshold for identifying outliers (e.g., z_score > 3)
threshold = 3

# Boolean mask for outliers
outlier_mask = (z_scores > threshold)
```

```python
# 3. Replace the outliers with the median value
# Calculate the median value for each column
medians = df.median()

# Replace outliers with the median value
df[outlier_mask] = np.where(outlier_mask, medians, df)

# Display the DataFrame after replacing outliers
print("DataFrame after replacing outliers with median value:")
print(df)
```
